```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv("Data/winequality-red.csv")

# Compute Z-scores for all numeric columns
z_scores = np.abs((df - df.mean()) / df.std())

# Remove rows where any column has a Z-score greater than 3
df = df[(z_scores < 3).all(axis=1)]
```

```python
# Count how many entries
nEntries = df['pH'].count()
print(f"Entries: {nEntries}")
```

```python
x = df['residual sugar'].dropna()
y = df['volatile acidity'].dropna()

```

```python
# Produce a scatter plot
plt.figure(figsize=(5,4))
plt.scatter(x, y, color="darkblue")
plt.title("Alcohol vs FSD")
plt.xlabel("Alcohol (%)")
plt.ylabel("FSD")
```

```python
# Compute the correlation matrix
corr_matrix = np.corrcoef(x, y)
print("Correlation matrix:")
print(corr_matrix)

# Extract correlation
corr_xy = corr_matrix[0,1]
print("Correlation betwwen x and y is ", corr_xy)
```

```python
# Produce a correlation matrix heatmap

import seaborn as sns
# Compute the correlation matrix
correlation_matrix = df.corr()

# Plot the correlation matrix as a heatmap with numbers
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="RdBu", linewidths=0.5, square=True, cbar=True)

# Title and layout adjustments
plt.title("Correlation Matrix Heatmap", fontsize=14)
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)

# Show the heatmap
plt.show()
```

```python
# Plot scatters

# Create a figure with three subplots in a row
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Scatter plot: Fixed Acidity vs. pH
axes[0].scatter(df['fixed acidity'], df['pH'], alpha=0.6, color="blue", edgecolors="black")
axes[0].set_xlabel("Fixed Acidity")
axes[0].set_ylabel("pH")
axes[0].set_title("Fixed Acidity vs. pH")

# Scatter plot: Fixed Acidity vs. Density
axes[1].scatter(df['fixed acidity'], df['density'], alpha=0.6, color="red", edgecolors="black")
axes[1].set_xlabel("Fixed Acidity")
axes[1].set_ylabel("Density")
axes[1].set_title("Fixed Acidity vs. Density")

# Scatter plot: Fixed Acidity vs. Citric Acid
axes[2].scatter(df['fixed acidity'], df['citric acid'], alpha=0.6, color="green", edgecolors="black")
axes[2].set_xlabel("Fixed Acidity")
axes[2].set_ylabel("Citric Acid")
axes[2].set_title("Fixed Acidity vs. Citric Acid")

# Adjust layout for better visibility
plt.tight_layout()

# Show the plot
plt.show()
```
