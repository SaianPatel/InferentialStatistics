```python
# Import require modules
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

# Set the number of times to roll each die
N = 10000

# Simulate rolling each die by taking N random integers from 1 to 6
dice_samples_1 = stats.randint(1, 7).rvs(N)
dice_samples_2 = stats.randint(1, 7).rvs(N)

# Sum (element-wise) the N samples of each die
dice_sum = dice_samples_1 + dice_samples_2

# Count how many occurences of each sum
counts = np.bincount(dice_sum)[1:]

# Print the counts of each sum
print("\nQuestion 1: Sum of two dice rolls")
for i, count in enumerate(counts, 1):
    print(f"Sum of {i}: {count} times")

# Get the total count of sums which are at least 9
counts_at_least_9 = np.sum(counts[8:])

# Display how many sums are at least 9
print(f"\nNumber of sums at least 9 in {N:.0f} attempts = {counts_at_least_9:.0f}")

# Calculate the estimated probability of the sum of 2 dice rolls being at least 9
p_at_least_9 = counts_at_least_9 / N

# Print the estimated probability
print(f"\nThe probability of the sum being at least 9 in {N:.0f} attempts = {p_at_least_9:.4f}")
```

```python
# Import require modules
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

# Set the number of times to repeat the experiment
nRepeats = 5000

# Set the number of times to roll each die
nRolls = 10000

# Set the theoretical probability that the sum is at least 9
p_theoretical = 5/18

# Initialise a list to store the estimated probability for each sample
P = []

# Repeat the simulation over nRepeats
for _ in range(nRepeats):

    # Roll each die 10000 times in each repeat
    dice_1 = stats.randint(1, 7).rvs(nRolls)
    dice_2 = stats.randint(1, 7).rvs(nRolls)

    # Sum the sides of the dice
    dice_sum_i = dice_1 + dice_2

    # Count how many times the sum is at least 9
    counts_i = np.bincount(dice_sum_i)[1:]

    # Calculate the probability of the sum being at least 9 and store it
    P.append(np.sum(counts_i[8:])/nRolls)


# Convert P to an array for convenience
P = np.array(P)

# Plot the distribution of probabilities
plt.hist(P, bins=30, density=True, alpha=0.75, edgecolor='black')

# Plot a vertical line at the theoretical probability
plt.axvline(x=p_theoretical, color='black', linestyle='dashed', linewidth=2, label="Theoretical Probability (5/18)")

# Define x-values to plot the theoretical normal distribution
x_vals = np.linspace(0.25, 0.31, 400)

# Compute the theoretical PDF for the probability:
theoretical_pdf = stats.norm.pdf(x_vals, loc= p_theoretical, scale=np.sqrt((p_theoretical)*(1 - p_theoretical)/nRolls))

plt.plot(x_vals, theoretical_pdf, 'r-', label='CLT Normal Approx')

plt.xlabel("Estimated Probability of Sum ≥ 9")
plt.ylabel("Density")
plt.title("Distribution of Estimated Probabilities Over Repeats")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the Bernoulli PMF
p = 0.55
x = [0, 1]  # Possible values of a Bernoulli variable
pmf = [1 - p, p]  # Probabilities for 0 and 1

plt.figure(figsize=(8,5))

# Plot vertical lines with circular markers at the probabilities
plt.vlines(x, ymin=0, ymax=pmf, colors="b", linestyles="dashed")

plt.scatter(x, pmf, color="b", zorder=3)

# Labels and title
plt.xlabel("X")
plt.ylabel("P(X)")
plt.title("PMF of a Bernoulli Random Variable (p = 0.55)")

# Show probability values next to markers
for i, v in zip(x, pmf):
    plt.text(i, v + 0.02, f"{v:.2f}", ha="center")

# Set limits for better visibility
plt.ylim(0, 1)

# Display the plot
plt.show()
```

```python
# Import required module
from scipy.stats import binom

# Set the probability of success, p, for the Binomial RV
p = 0.9

# Set the number of trials, n, for the Binomial RV
n = 30

# Set the value of Y of interest
y = 27

# Calculate Pr(Y>= 28) (which is 1 - Pr(Y<=27))
p_at_least_28 = 1 - binom.cdf(y, n, p)

# Print the result
print(f"\nPr(Y>28) for Y ~ B(30, 0.9) is {p_at_least_28:.3f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the continuous uniform distribution parameters
a, b = 2, 5  # Range of the distribution
x = np.linspace(a - 1, b + 1, 1000)  # Range for plotting
pdf = np.where((x >= a) & (x <= b), 1 / (b - a), 0)  # PDF of U(2,5)

plt.figure(figsize=(8,5))

# Plot the PDF
plt.plot(x, pdf, 'b-', lw=2, label="PDF of U(2,5)")
# plt.axvline(a, color='black', linestyle='dashed', alpha=0.7)  # Vertical line at a
# plt.axvline(b, color='black', linestyle='dashed', alpha=0.7)  # Vertical line at b
plt.axhline(1/3, color='black', linestyle='dashed', alpha=0.7)  # Vertical line at b
plt.fill_between(x, pdf, alpha=0.2, color='blue', where=(x >= a) & (x <= b))  # Shade area under PDF

# Labels and title
plt.xlabel("W")
plt.ylabel("f(W)")
plt.title("PDF of a Continuous Uniform Distribution U(2,5)")
# plt.legend()

# Display the plot
plt.show()
```

```python
# Import required module
from scipy.stats import norm

# Set the mean of the Normal Distribution
mu = 50

# Set the standard deviation of the Normal Distribution
sigma = 4

# Calculate Pr(S ≤ 45) using the cdf
p_45 = norm(mu, sigma).cdf(45)

# Print the result
print(f"\nPr(S ≤ 45) for S ~ N(50, 4^2) is {p_45:.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Given parameters
mu = 50  # Mean
sigma = 4  # Standard deviation

# Generate x values for the normal distribution
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)
pdf = norm.pdf(x, mu, sigma)  # Normal PDF

# Calculate Pr(S ≤ 45) using the CDF
p_45 = norm.cdf(45, mu, sigma)

plt.figure(figsize=(8,5))

# Plot the normal distribution
plt.plot(x, pdf, 'r-', lw=2, label="Normal Distribution N(50, 4^2)")

# Shade the area for P(S ≤ 45)
x_shade = np.linspace(mu - 4 * sigma, 45, 500)
plt.fill_between(x_shade, norm.pdf(x_shade, mu, sigma), alpha=0.3, color='blue', label="P(S ≤ 45)")

# Labels and title
plt.xlabel("S")
plt.ylabel("Density")
plt.title("Normal Distribution and P(S ≤ 45)")
plt.legend()

# Show probability value on plot
plt.text(45.2, 0.02, f"P(S ≤ 45) = {p_45:.4f}", fontsize=12, color="blue")

# Display the plot
plt.show()

# Print the probability value
print(f"\nPr(S ≤ 45) for S ~ N(50, 4^2) is {p_45:.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Given parameters
mu = 50  # Mean
sigma = 4  # Standard deviation

# Generate x values for the normal distribution
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)
pdf = norm.pdf(x, mu, sigma)  # Normal PDF

# Calculate Pr(48 ≤ S ≤ 55) using the CDF
p_48_55 = norm.cdf(55, mu, sigma) - norm.cdf(48, mu, sigma)

# Initialise the figure
plt.figure(figsize=(8,5))

# Plot the normal distribution
plt.plot(x, pdf, 'r-', lw=2, label="Normal Distribution N(50, 4^2)")

# Shade the area for P(48 ≤ S ≤ 55)
x_shade = np.linspace(48, 55, 500)
plt.fill_between(x_shade, norm.pdf(x_shade, mu, sigma), alpha=0.3, color='blue', label="P(48 ≤ S ≤ 55)")

# Labels and title
plt.xlabel("S")
plt.ylabel("Density")
plt.title("Normal Distribution and P(48 ≤ S ≤ 55)")
# plt.legend()

# Show probability value on plot
plt.text(52.5, 0.09, f"P(48 ≤ S ≤ 55) = {p_48_55:.4f}", fontsize=12, color="blue")

# Display the plot
plt.show()

# Print the probability value
print(f"\nPr(48 ≤ S ≤ 55) for S ~ N(50, 4^2) is {p_48_55:.4f}")
```

```python
# Import required modules
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

# Set the number of repeats
nRepeats = 2000

# Set the sample size
nSamples = 100

# Set p for the Bernoulli RV
p = 0.4

# Set the variance for the Bernoulli RV
v = p * (1-p)

# Initialise a list to store the sample means
sampleMeans = []

# Repeat the sampling process over nRepeats
for _ in range(nRepeats):

    # Draw a random number between 0 and 1 and compare to p
    samples = (np.random.rand(nSamples) < p).astype(int)

    # Get the mean of the sample and store it
    sampleMeans.append(samples.mean())

# Convert P to an array for convenience
sampleMeans = np.array(sampleMeans)

# Initialise the figure
plt.figure(figsize=(8,5))

# Plot the distribution of probabilities
plt.hist(sampleMeans, bins=30, density=True, alpha=0.75, edgecolor='black')

# Plot a vertical line at the theoretical probability
plt.axvline(x= p, color='black', linestyle='dashed', linewidth=2, label="Theoretical Mean of 0.4")

# Define x-values to plot the theoretical normal distribution
x_vals = np.linspace(0.2, 0.6, 400)

# Compute the theoretical PDF for the sample means
# N(p, sqrt(v/n))
theoretical_pdf = stats.norm.pdf(x_vals, loc=p, scale=np.sqrt(v/nSamples))

# Overlay the theoretical PDF for the sample means
plt.plot(x_vals, theoretical_pdf, 'r-', label='CLT Normal Approx')

plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.title("Distribution of Sample Means Over Repeats")
plt.legend()
plt.show()
```
