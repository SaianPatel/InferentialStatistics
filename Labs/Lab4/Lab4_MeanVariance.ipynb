```python
# ===========================================================
# 1. SETUP
# ===========================================================
# We'll import:
#  - numpy for numerical computations (arrays, mean, var, std)
#  - matplotlib.pyplot for plotting histograms
#  - from scipy.stats, the distribution classes:
#      uniform, binom, bernoulli, norm
#    Each of these classes has .mean(), .var(), .std() methods we can use.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, binom, bernoulli, norm

# We can set a random seed for reproducibility.
np.random.seed(42)

# If needed in Jupyter, ensure inline plotting:
%matplotlib inline

print("Setup complete. Libraries imported and random seed set.")
```

```python
# ===========================================================
# 2. UNIFORM DISTRIBUTION
# ===========================================================
# In this section:
#   1) We create a uniform distribution object with Uniform(a, b).
#   2) We use .mean(), .var(), .std() on that distribution object
#      to get the theoretical mean, variance, and std dev.
#   3) We sample data (empirical) using .rvs().
#   4) We compare the empirical statistics with the theoretical values.

# Let's define our uniform distribution on [0, 10].
a, b = 0, 10

# Create a "frozen" distribution object by calling uniform(...) with parameters
dist_uniform = uniform(loc=a, scale=(b - a))

# Now we can call dist_uniform.mean(), dist_uniform.var(), dist_uniform.std()
# to get the "true" or theoretical values for this distribution.
theoretical_mean_uni = dist_uniform.mean()
theoretical_var_uni = dist_uniform.var()
theoretical_std_uni = dist_uniform.std()

print(f"Uniform Distribution on [{a}, {b}]")
print("Theoretical mean (dist_uniform.mean()):", theoretical_mean_uni)
print("Theoretical variance (dist_uniform.var()):", theoretical_var_uni)
print("Theoretical std dev (dist_uniform.std()):", theoretical_std_uni)

# We now draw samples from this distribution using .rvs().
N_uni = 5000
data_uniform = dist_uniform.rvs(size=N_uni)

# Compute the empirical values from the sample
sample_mean_uni = np.mean(data_uniform)
sample_var_uni  = np.var(data_uniform, ddof=1)  # ddof=1 gives sample variance
sample_std_uni  = np.std(data_uniform, ddof=1)  # ddof=1 gives sample std dev

print("\nFrom sampling N=5000 data points:")
print("Sample mean:", sample_mean_uni)
print("Sample variance:", sample_var_uni)
print("Sample std dev:", sample_std_uni)

# We can make a quick histogram to visualize the data
plt.figure(figsize=(8,5))
plt.hist(data_uniform, bins=30, density=True, alpha=0.6, edgecolor='black')
plt.title("Histogram of Uniform(0, 10) Samples")
plt.grid(alpha = 0.3)
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# ===========================================================
# 3. BINOMIAL DISTRIBUTION
# ===========================================================
# Similarly, we do the steps for a Binomial(n, p) distribution.
#   1) Create a frozen distribution object: binom(n, p)
#   2) Get theoretical stats from .mean(), .var(), .std()
#   3) Sample with .rvs() to get empirical data
#   4) Compare

n_binom, p_binom = 20, 0.3

# Create the distribution object
dist_binom = binom(n=n_binom, p=p_binom)

# Get theoretical stats
theoretical_mean_binom = dist_binom.mean()
theoretical_var_binom  = dist_binom.var()
theoretical_std_binom  = dist_binom.std()

print(f"\nBinomial Distribution with n={n_binom}, p={p_binom}")
print("Theoretical mean (dist_binom.mean()):", theoretical_mean_binom)
print("Theoretical variance (dist_binom.var()):", theoretical_var_binom)
print("Theoretical std dev (dist_binom.std()):", theoretical_std_binom)

# Generate samples
N_binom = 5000
data_binom = dist_binom.rvs(size=N_binom)

# Empirical stats
sample_mean_binom = np.mean(data_binom)
sample_var_binom  = np.var(data_binom, ddof=1)
sample_std_binom  = np.std(data_binom, ddof=1)

print("\nFrom sampling N=5000 data points:")
print("Sample mean:", sample_mean_binom)
print("Sample variance:", sample_var_binom)
print("Sample std dev:", sample_std_binom)

# Quick histogram (discrete distribution)
bins_range = range(0, n_binom + 2)
plt.figure(figsize=(8,5))
plt.hist(data_binom, bins=bins_range, density=True, alpha=0.6, edgecolor='black')
plt.grid(alpha = 0.3)
plt.title(f"Histogram of Binomial(n={n_binom}, p={p_binom}) Samples")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

```python
# ===========================================================
# 4. BERNOULLI DISTRIBUTION
# ===========================================================
# Bernoulli(p) is just a special case of the binomial with n=1,
# but scipy.stats provides a bernoulli distribution class.

p_bern = 0.7

# Create the distribution object
dist_bern = bernoulli(p=p_bern)

# Theoretical stats
theoretical_mean_bern = dist_bern.mean()
theoretical_var_bern  = dist_bern.var()
theoretical_std_bern  = dist_bern.std()

print(f"\nBernoulli Distribution with p={p_bern}")
print("Theoretical mean (dist_bern.mean()):", theoretical_mean_bern)
print("Theoretical variance (dist_bern.var()):", theoretical_var_bern)
print("Theoretical std dev (dist_bern.std()):", theoretical_std_bern)

# Generate samples
N_bern = 5000
data_bern = dist_bern.rvs(size=N_bern)

# Empirical stats
sample_mean_bern = np.mean(data_bern)
sample_var_bern  = np.var(data_bern, ddof=1)
sample_std_bern  = np.std(data_bern, ddof=1)

print("\nFrom sampling N=5000 data points:")
print("Sample mean:", sample_mean_bern)
print("Sample variance:", sample_var_bern)
print("Sample std dev:", sample_std_bern)

# Quick histogram (0 or 1)
plt.figure(figsize=(5,4))
plt.hist(data_bern, bins=[-0.5, 0.5, 1.5], density=True, alpha=0.6, edgecolor='black')
plt.title(f"Histogram of Bernoulli(p={p_bern}) Samples")
plt.xlabel("Value (0 or 1)")
plt.ylabel("Probability")
plt.grid(alpha = 0.3)
plt.xticks([0,1])
plt.show()
```

```python
# ===========================================================
# 5. GAUSSIAN (NORMAL) DISTRIBUTION
# ===========================================================
# We'll do the same for a Normal distribution.

mu_norm, sigma_norm = 5.0, 1.5

# Create the distribution object
dist_norm = norm(loc=mu_norm, scale=sigma_norm)

# Theoretical stats
theoretical_mean_norm = dist_norm.mean()
theoretical_var_norm  = dist_norm.var()
theoretical_std_norm  = dist_norm.std()

print(f"\nNormal (Gaussian) Distribution with mean={mu_norm}, std={sigma_norm}")
print("Theoretical mean (dist_norm.mean()):", theoretical_mean_norm)
print("Theoretical variance (dist_norm.var()):", theoretical_var_norm)
print("Theoretical std dev (dist_norm.std()):", theoretical_std_norm)

# Generate samples
N_norm = 5000
data_norm = dist_norm.rvs(size=N_norm)

# Empirical stats
sample_mean_norm = np.mean(data_norm)
sample_var_norm  = np.var(data_norm, ddof=1)
sample_std_norm  = np.std(data_norm, ddof=1)

print("\nFrom sampling N=5000 data points:")
print("Sample mean:", sample_mean_norm)
print("Sample variance:", sample_var_norm)
print("Sample std dev:", sample_std_norm)

# Histogram of the sample
plt.figure(figsize=(8,5))
plt.hist(data_norm, bins=30, density=True, alpha=0.6, edgecolor='black')
plt.title(f"Histogram of Normal(mean={mu_norm}, std={sigma_norm}) Samples")
plt.grid(alpha = 0.3)
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```
