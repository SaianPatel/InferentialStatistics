```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

# Exercise 1: Binomial PMF
n1, p1, k1 = 12, 0.5, 5
prob1 = stats.binom.pmf(k1, n1, p1)
print(f"Exercise 1: Probability of 5 heads in 12 flips = {prob1:.4f}")
```

```python

# Exercise 2: Binomial CDF
n2, p2, k2 = 15, 0.08, 2
prob2 = stats.binom.cdf(k2, n2, p2)
print(f"\nExercise 2: Probability of ≤2 defects in 15 trials = {prob2:.4f}")
```

```python

# Exercise 3: Uniform Sampling (Dice)
dice_samples = stats.randint(1, 7).rvs(5000)
mean3 = np.mean(dice_samples)
print(f"\nExercise 3: Empirical mean of 5000 dice rolls = {mean3:.2f}")
```

```python

# Exercise 4: Uniform Probability
a4, b4 = 0, 20
prob4 = stats.uniform.cdf(16, a4, b4-a4) - stats.uniform.cdf(4, a4, b4-a4)
print(f"\nExercise 4: P(4 ≤ X ≤ 16) = {prob4:.2f}")
```

```python

# Exercise 5: Gaussian PDF
mu5, sigma5, x5 = 10, 3, 8
pdf5 = stats.norm(mu5, sigma5).pdf(x5)
print(f"\nExercise 5: PDF at x=8 = {pdf5:.4f}")
```

```python

# Exercise 6: Gaussian CDF
mu6, sigma6, x6 = 12, 4, 15
cdf6 = stats.norm(mu6, sigma6).cdf(x6)
print(f"\nExercise 6: P(X ≤ 15) = {cdf6:.4f}")
```

```python

# Exercise 7: Binomial Sampling
n7, p7 = 8, 0.4
samples7 = stats.binom.rvs(n7, p7, size=2000)
plt.hist(samples7, bins=range(n7+2), density=True, edgecolor='black')
plt.title("Exercise 7: Binomial Samples (n=8, p=0.4)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python

# Exercise 8: Uniform vs. Gaussian
x_unif = np.linspace(0, 10, 100)
pdf_unif = stats.uniform(0, 10).pdf(x_unif)

x_gauss = np.linspace(-5, 15, 100)
pdf_gauss = stats.norm(5, 2).pdf(x_gauss)

plt.plot(x_unif, pdf_unif, label="Uniform (0-10)")
plt.plot(x_gauss, pdf_gauss, label="Gaussian (μ=5, σ=2)")
plt.title("Exercise 8: Distribution Comparison")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python

# Exercise 9: Empirical Rule Validation
mu9, sigma9 = 25, 5
samples9 = stats.norm.rvs(mu9, sigma9, size=10000)
within_2sigma = np.sum((samples9 >= mu9-2*sigma9) & (samples9 <= mu9+2*sigma9))
prop9 = within_2sigma / 10000
print(f"\nExercise 9: Proportion within μ±2σ = {prop9:.3f}")
```

```python

# Exercise 10: Dice Roll Frequency
dice_rolls = stats.randint(1, 7).rvs(10000)
counts = np.bincount(dice_rolls)[1:]
print("\nExercise 10: Dice Roll Counts:")
for i, count in enumerate(counts, 1):
    print(f"Number {i}: {count} times")
```
